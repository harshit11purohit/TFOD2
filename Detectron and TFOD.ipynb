```python
# What is Detectron2?
# A PyTorch-based library for:
# - Object Detection
# - Instance Segmentation
# - Keypoint Detection

# Built by Facebook AI (Meta)
```

```python
# Core Architecture:

# 1. Backbone
# Extracts features from the image
# Examples:
# - ResNet50
# - ResNet101
# Image → Feature maps

# 2. RPN (Region Proposal Network)
# Finds possible object regions
# Feature maps → candidate boxes
# - Generates bounding box proposals
# - Filters likely object areas

# 3. ROI Heads
# Final decision stage
# Proposals → classify + refine boxes
# - Classifies object
# - Improves bounding box accuracy
```

```python
# TensorFlow Object Detection API
# A TensorFlow-based framework for training, evaluating, and deploying object detection models. It provides ready-to-use architectures, training pipelines, and evaluation tools.
```

```python
# Core components
# 1. Model architectures

# Available through a model zoo (pretrained checkpoints + configs):
# SSD (Single Shot Detector)
# Faster R-CNN
# EfficientDet
# CenterNet

# Each model defines:
# Backbone (feature extractor)
# Detection head (classification + box regression)
```

```python
# 2. Pipeline configuration (central concept)

# Everything is controlled by a single pipeline.config file.

# This file defines:

# Model type
# Dataset paths
# Training parameters
# Evaluation settings
```
